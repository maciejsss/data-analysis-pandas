```python
retail.assign(
    year = retail['date'].dt.year
)
```

```python
retail = retail.assign(
    birthday = pd.to_datetime('1950-01-01')
)
```

```python
retail
```

```python
retail = retail.assign(
    age = (retail['date'] - retail['birthday']).dt.days / 365.25
)

retail
```

```python
retail.assign(
    retirement_date = retail['date'] + (pd.to_timedelta(365.25,unit='D') * 20)
)
```

```python
retail_agg = retail.groupby(['date']).agg({
    'sales': 'sum'
}).round().iloc[1:10]
```

```python
retail_agg
```

```python
retail_agg.iloc[1] = np.NaN
retail_agg.iloc[4:6] = np.nan
```

```python
retail_agg
```

```python
retail_agg.fillna(retail_agg['sales'].mean())
```

```python
retail_agg.ffill()
```

```python
retail_agg = retail_agg.interpolate().round()
```

```python
retail_agg
```

```python
retail_agg.assign(
    yesterday_sales = retail_agg['sales'].shift()
)
```

```python
retail_agg.assign(
    ratio = retail_agg['sales'] / retail_agg['sales'].shift()
)
```

```python
retail_agg.assign(
    diff = retail_agg['sales'] - retail_agg['sales'].shift()
)
```

```python
retail_agg.assign(
    diff = retail_agg['sales'].diff()
)
```

```python
retail.groupby([retail['date'].dt.year, retail['date'].dt.month]).agg({
    'sales': 'sum'
}).round(2)
```

```python
retail.set_index('date').loc[:,['sales']].resample('M').sum().round(2)
```

```python
retail_daily = retail.set_index('date').loc[:,'sales'].resample('D').sum().round(2)
```

```python
retail_daily
```
