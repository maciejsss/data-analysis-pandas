```python
small_retail = retail[['store_nbr','family','sales','onpromotion']]
```

```python
small_retail.groupby(['store_nbr','family']).agg('sum')
```

```python
small_retail.groupby(['store_nbr','family']).agg(['sum','mean'])
```

```python
(
    small_retail
    .groupby(['store_nbr','family'])
    .agg(
        {
            'sales':['sum','mean'],
            'onpromotion':['min','max']
        }
    )
 )
```

```python
(
    small_retail
    .groupby(['store_nbr','family'])
    .agg(
        sales_sum = ('sales','sum'),
        sales_avg = ('sales', 'mean'),
        onpromotion_max = ('onpromotion','max')
    )
 )
```

```python
(
    premier_league
    .groupby(['season','HomeTeam'], as_index=False)
    .agg(
        {
            'HomeGoals':['sum','mean'],
            'AwayGoals':['sum','mean']
        }
    )
)
```

```python
pm = (
    
    premier_league
    .query('HomeTeam in ["Arsenal","Chelsea","Everton"] and season in  ["2008/2009","2009/2010","2010/2011"]')
    .pivot_table(
        index='HomeTeam',
        columns='season',
        values='HomeGoals',
        aggfunc='sum'
    )
)
```

```python
(
    premier_league
    .query('HomeTeam in ["Arsenal","Chelsea","Everton"] and season in  ["2008/2009","2009/2010","2010/2011"]')
    .pivot_table(
        index='HomeTeam',
        columns='season',
        values='HomeGoals',
        aggfunc=('sum','mean')
    )
)
```

```python
(
    premier_league
    .query('HomeTeam in ["Arsenal","Chelsea","Everton"] and season in  ["2008/2009","2009/2010","2010/2011"]')
    .pivot_table(
        index='HomeTeam',
        columns='season',
        values=['HomeGoals','AwayGoals'],
        aggfunc=({'HomeGoals':['sum','mean'], 'AwayGoals':['max']})
    ).style.background_gradient(cmap='RdYlGn',axis=None)
)
```

```python
(
    premier_league
    .query('HomeTeam in ["Arsenal","Chelsea","Everton"] and season in  ["2008/2009","2009/2010","2010/2011"]')
    .pivot_table(
        index='HomeTeam',
        columns='season',
        values=['HomeGoals','AwayGoals'],
        aggfunc=({'HomeGoals':['sum','mean'], 'AwayGoals':['max']})
    ).style.background_gradient(cmap='RdYlGn',axis=1)
)
```
