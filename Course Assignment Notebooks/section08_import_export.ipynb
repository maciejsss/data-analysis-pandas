```python
# expand on the code below...
columns = ['Date','Store_Number','Transaction_Count']

transactions = pd.read_csv(
    "../retail/transactions.csv",
    names=columns,
    skiprows=[0],
    parse_dates=['Date']
)
```

```python
transactions.assign(
    target_pct=transactions["Transaction_Count"] / 2500,
    met_target=(transactions["Transaction_Count"] / 2500) >= 1,
    bonus_payable=((transactions["Transaction_Count"] / 2500) >= 1) * 100,
    month=transactions["Date"].dt.month,
    day_of_week=transactions["Date"].dt.dayofweek,
)
```

# Assignment 2: Write to Excel Sheets

Write the data in the transactions dataframe you created above into an Excel workbook.

Write out a separate sheet for each year of the data.

If you prefer, you can write each year of data to a separate csv file.

```python
transactions.index = transactions['Date']
```

```python
transactions = transactions.drop(columns='Date')
```

```python
with pd.ExcelWriter('../retail/transactions.xlsx') as writer:
    transactions.loc['2013'].to_excel(writer, sheet_name='2013'),
    transactions.loc['2014'].to_excel(writer, sheet_name='2014'),
    transactions.loc['2015'].to_excel(writer, sheet_name='2015'),
    transactions.loc['2016'].to_excel(writer, sheet_name='2016'),
    transactions.loc['2017'].to_excel(writer, sheet_name='2017')
```

```python
# range = [*range(2013,2014)]
# years = [str(element) for element in range]

years = ['2013', '2014']


with pd.ExcelWriter('../retail/transactions.xlsx') as writer:
    for year in years:
        transactions.loc[year].to_excel(writer, sheet_name=year)

```
